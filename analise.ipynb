```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Lendo o arquivo que você acabou de gera
df = pd.read_csv('clientes_telecom.csv')

# Definindo o estilo dos gráficos
sns.set_theme(style="whitegrid")


# Gráfico 1: Tipo de Contrato vs Churn
plt.figure(figsize=(101, 5))
sns.countplot(data=df, x='TipoContrato', hue='Churn', palette='viridis')
plt.title('Cancelamentos (Churn) por Tipo de Contrato', fontsize=14, fontweight='bold')
plt.xlabel('Tipo de Contrato', fontsize=12)
plt.ylabel('Quantidade de Clientes', fontsize=12)
plt.legend(title='Cancelou? (Churn)')
plt.show()

# Gráfico 2: Reclamações no Suporte vs Churn
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Churn', y='ReclamacoesSuporte', palette='mako')
plt.title('Numero de Reclamacoes no Suporte vs Status do Cliente', fontsize=14, fontweight='bold')
plt.xlabel('Cancelou o Serviço? (Churn)', fontsize=12)
plt.ylabel('Quantidade de Reclamacoes', fontsize=12)
plt.show()
```

```python
import pandas as pd 
import numpy as np 

# Definindo uma semente para os dados serem sempre iguais
np.random.seed(42)
n_clientes = 1000

# Criando dados fictícios de clientes
dados = {
    'ID_Cliente': [f'ID-{i:04d}' for i in range(1, n_clientes + 1)],
    'TipoContrato': np.random.choice(['Mensal', 'Anual', 'Bienal'], n_clientes, p=[0.5, 0.3, 0.2]),
    'ValorMensal': np.round(np.random.uniform(50.0, 150.0, n_clientes), 2),
    'MesesNaEmpresa': np.random.randint(1, 72, n_clientes),
    'ReclamacoesSuporte': np.random.poisson(1.5, n_clientes),
}

df = pd.DataFrame(dados)

# Regra lógica para o Churn (simulando comportamento real)
# Clientes com contrato mensal e muitas reclamações têm mais chance de cancelar (Churn = 'Sim')
def calcular_churn(row):
    score = 0
    if row['TipoContrato'] == 'Mensal': score += 40
    if row['ReclamacoesSuporte'] > 3: score += 40
    if row['MesesNaEmpresa'] < 12: score += 20
    
    return 'Sim' if score + np.random.randint(-20, 20) > 50 else 'Não'

df['Churn'] = df.apply(calcular_churn, axis=1)

# Salvando em um arquivo CSV para usarmos depois
df.to_csv('clientes_telecom.csv', index=False)
print("Base de dados 'clientes_telecom.csv' gerada com sucesso!")
```
